#Seller provide Max order to Snapdeal

```python
sdf=df.groupby(['sellercode']).agg(
    seller=('sellercode','count')
)
sdf=sdf.sort_values(by='seller',ascending=False)
sdf=sdf.reset_index()
sdf
```

#Find Category Under Danger Good

```python
sc=df.groupby('subcategory_url').agg(
    DGcat=('subcategory_url','count')
)
sc=sc.reset_index()
sc=sc.sort_values(by='DGcat',ascending=False)
sc
```
